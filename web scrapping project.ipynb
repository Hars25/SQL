```python
url='https://www.ebay.com/sch/i.html?_from=R40&_trksid=p2334524.m570.l1313&_nkw=rolex+luxury+watches&_sacat=0&LH_TitleDesc=0&_odkw=rolex+latestt+watches&_osacat=0'
```

```python
page=requests.get(url)
```

```python
print(page)
```

```python
soup=BeautifulSoup(page.content)
```

```python
soup.prettify()
```

```python
watch=soup.find(id="srp-river-main")
```

```python
print(watch)
```

```python
items=watch.find_all(class_="s-item__info clearfix")
```

```python
print(items)
```

```python
w=items[3]
w
print(w.prettify())
```

```python
desc=w.find(class_="s-item__title").get_text()
print(desc)
```

```python
price=w.find(class_="s-item__price").get_text()
print(price)
```

```python
location=w.find(class_="s-item__location s-item__itemLocation").get_text()
print(location)
```

```python
shippingcharge=w.find(class_="s-item__shipping s-item__logisticsCost").get_text()
print(shippingcharge)
```

```python
itemssold=w.find(class_="s-item__dynamic s-item__additionalItemHotness").get_text()
print(itemssold)
```

```python
location_tags=watch.select(".s-item__info clearfix .s-item__location s-item__itemLocation")
print(location_tags)
```

```python
itemssold_tags=watch.select(".srp-river-main .s-item__dynamic s-item__additionalItemHotness")
print(itemssold_tags)
```

```python
shippingcharge_tags=watch.select(".s-item__details clearfix .s-item__detail--primary .s-item__shipping s-item__logisticsCost")
print(shippingcharge_tags)
```

```python
desc_tags=watch.select(".srp-river-main .s-item__title")
print(desc_tags)
```

```python
desc=[i.get_text() for i in desc_tags]
print(desc)
```

```python
price_tags=watch.select(".srp-river-main .s-item__price")
print(price_tags)
```

```python
price=[i.get_text() for i in price_tags]
print(price)
```
